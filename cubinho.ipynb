```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np

# %matplotlib inline

#plt.rcParams['axes.unicode_minus'] = False
#plt.rcParams['font.sans-serif'] = ['SimHei']


def plot_linear_cube(x, y, z, dx, dy, dz, color='blue'):
    fig = plt.figure()
    ax = Axes3D(fig)
    xx = [x, x, x+dx, x+dx, x]
    yy = [y, y+dy, y+dy, y, y]
    kwargs = {'alpha': 1, 'color': color}
    color2 = 'red'
    kw = {'alpha':1,'color':color2}
    ax.plot3D(xx, yy, [z], **kwargs)   #face inferior
    ax.plot3D(xx, yy, [z+dz], **kwargs) #face superior
    ax.plot3D([x, x], [y, y], [z, z+dz], **kwargs)
    ax.plot3D([x, x], [y+dy, y+dy], [z, z+dz], **kwargs)
    ax.plot3D([x+dx, x+dx], [y+dy, y+dy], [z, z+dz], **kwargs)
    ax.plot3D([x+dx, x+dx], [y, y], [z, z+dz], **kwargs)
    ax.plot3D([x,x+dx],[y,y],[z,z+dz],**kw)
    plt.title('Cube')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()


plot_linear_cube(0, 0, 0, 1, 1, 1)
plot_linear_cube(0, 0, 0,-1,1,1)
    #plot_opaque_cube()
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

#Definicao dos 8 vertices

def decomp(x,y,z,F,cor = 'blue'):
    #Vertices
    V = np.array([[x,0,0],
                  [x,y,0],
                  [0,y,0],
                  [0,0,0],
                  [x,0,z],
                  [x,y,z],
                  [0,y,z],
                  [0,0,z]])

    #Aplicação do Gradiente
    Vnew = []
    for i in range(len(V)):
      A = []
      A = np.dot(F,V[i])
      Vnew.append(A)

    #Novos Vertices
    v1 = Vnew[0]
    v2 = Vnew[1]
    v3 = Vnew[2]
    v4 = Vnew[3]
    v5 = Vnew[4]
    v6 = Vnew[5]
    v7 = Vnew[6]
    v8 = Vnew[7]

    #Definicao das arestas
    #Face inferior
    a1 = [[v1[0],v2[0]],[v1[1],v2[1]],[v1[2],v2[2]]]
    a2 = [[v2[0],v3[0]],[v2[1],v3[1]],[v2[2],v3[2]]]
    a3 = [[v3[0],v4[0]],[v3[1],v4[1]],[v3[2],v4[2]]]
    a4 = [[v4[0],v1[0]],[v4[1],v1[1]],[v4[2],v1[2]]]

    #arestas na vertical
    a5 = [[v1[0],v5[0]],[v1[1],v5[1]],[v1[2],v5[2]]]
    a6 = [[v2[0],v6[0]],[v2[1],v6[1]],[v2[2],v6[2]]]
    a7 = [[v3[0],v7[0]],[v3[1],v7[1]],[v3[2],v7[2]]]
    a8 = [[v4[0],v8[0]],[v4[1],v8[1]],[v4[2],v8[2]]]

    #face superior
    a9 = [[v5[0],v6[0]],[v5[1],v6[1]],[v5[2],v6[2]]]
    a10 = [[v6[0],v7[0]],[v6[1],v7[1]],[v6[2],v7[2]]]
    a11 = [[v7[0],v8[0]],[v7[1],v8[1]],[v7[2],v8[2]]]
    a12 = [[v8[0],v5[0]],[v8[1],v5[1]],[v8[2],v5[2]]]

    fig = plt.figure()
    ax = Axes3D(fig)
    kw = {'alpha': 2, 'color': cor}

    ax.plot3D(a1[0],a1[1],a1[2],**kw)
    ax.plot3D(a2[0],a2[1],a2[2],**kw)
    ax.plot3D(a3[0],a3[1],a3[2],**kw)
    ax.plot3D(a4[0],a4[1],a4[2],**kw)
    ax.plot3D(a5[0],a5[1],a5[2],**kw)
    ax.plot3D(a6[0],a6[1],a6[2],**kw)
    ax.plot3D(a7[0],a7[1],a7[2],**kw)
    ax.plot3D(a8[0],a8[1],a8[2],**kw)
    ax.plot3D(a9[0],a9[1],a9[2],**kw)
    ax.plot3D(a10[0],a10[1],a10[2],**kw)
    ax.plot3D(a11[0],a11[1],a11[2],**kw)
    ax.plot3D(a12[0],a12[1],a12[2],**kw)

    plt.title('Cubo')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
x = 1;
y = 1;
z = 1;

F = np.array([[1,0,0],[0,1,0],[0,0,1]])
F1 = np.array([[4,5,6],[0.5,1,0],[-0.5,0,1]])

decomp(x,y,z,F)
decomp(x,y,z,F1,'red')
```
