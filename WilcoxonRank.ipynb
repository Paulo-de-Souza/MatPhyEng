## Exemplo criado pelo chatgpt

```python

# Suponha que você tenha dois conjuntos de dados
x = np.array([1.2, 2.4, 2.8, 3.6, 4.0, 5.1, 5.9, 6.3, 7.2, 8.1])
y = np.array([1.5, 2.3, 2.7, 3.7, 4.2, 5.0, 5.8, 6.5, 7.1, 8.3])

# Calcule as diferenças
d = x - y

# Calcule os ranks das diferenças absolutas
ranks = rankdata(abs(d))

# Determine os ranks positivos e negativos
ranks_pos = np.sum(ranks[d > 0])
ranks_neg = np.sum(ranks[d < 0])

# Realize o teste de Wilcoxon
stat, p_value = wilcoxon(x, y)

print("Ranks positivos:", ranks_pos)
print("Ranks negativos:", ranks_neg)
print("Estatística do teste:", stat)
print("p-valor:", p_value)
```

```python
# retirando a primeira coluna
df1 = df.drop(df.columns[0], axis=1)

# Função para normalizar uma linha
def normalize(row):
    return (row - row.min()) / (row.max() - row.min())

# Aplica a função de normalização a cada linha do DataFrame
df_normalized = df1.apply(normalize, axis=1)
df_normalized
```

```python
SaDE_norm = df_normalized['SaDE'].to_numpy()
PSO_norm = df_normalized['PSO'].to_numpy()

# Calcule as diferenças
d1_norm = SaDE_norm - PSO_norm
# d2 = SaDE - CHC

# Calcule os ranks das diferenças absolutas
ranks = rankdata(abs(d1_norm))

# Determine os ranks positivos e negativos
ranks_pos = np.sum(ranks[d1_norm > 0])
ranks_neg = np.sum(ranks[d1_norm < 0])

# Realize o teste de Wilcoxon
stat, p_value = wilcoxon(SaDE_norm, PSO_norm)

print("Ranks positivos:", ranks_pos)
print("Ranks negativos:", ranks_neg)
print("Estatística do teste:", stat)
print("p-valor:", p_value)
```
