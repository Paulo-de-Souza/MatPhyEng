```python
from sympy import *
```

```python
x1,x2,x3,X1,X2,X3,t = symbols('x1 x2 x3 X1 X2 X3 t') 
```

```python
#Descricao do movimento
x1 = X1*3 + X1*X2*2*cos(t)
x2 = X2*4*(t**2)
x3 = X3*X2 + X1*t

#Coordenadas do gradiente de deformações
F11 = diff(x1,X1)
F12 = diff(x1,X2)
F13 = diff(x1,X3)
F21 = diff(x2,X1)
F22 = diff(x2,X2)
F23 = diff(x2,X3)
F31 = diff(x3,X1)
F32 = diff(x3,X2)
F33 = diff(x3,X3)
```

```python
#Matriz do Gradiente de Deformacoes
F = Matrix([[F11,F12,F13],
           [F21,F22,F23],
           [F31,F32,F33]])
F
```

```python
#Jacobiano 
J = F.det()
J
```

```python
#Descrição Material
p = 2*x1 + 2*x2 + x3*t
p
```

```python
#Derivada Material
dpdt = diff(p,t)
dpdt
```

```python
#Componente da Velocidade
v1 = diff(x1,t)
v2 = diff(x2,t)
v3 = diff(x3,t)

v = Matrix([v1,v2,v3])
v
```

```python
#Componente de Aceleração
a1 = diff(v1,t)
a2 = diff(v2,t)
a3 = diff(v3,t)

a = Matrix([a1,a2,a3])
a
```

```python
from sympy.plotting import plot3d

ac1 = a[0]
ac1 = ac1.subs(X1,-1)
plot3d(ac1,(t,-2*pi,2*pi),(X2,-2,4),title = 'Aceleração 1',xlabel = 'Tempo',ylabel = 'X2')
```
