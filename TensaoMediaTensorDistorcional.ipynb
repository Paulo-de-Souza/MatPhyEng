```python
import numpy as np
```

```python
#Tensor de Tensões
s = np.array([[50,30,20],
             [30,-20,-10],
              [20,-10,10]])
s
```

```python
#Tensao Média
def tensaomedia(s):
  sm = (1/3)*(s[0,0]+s[1,1]+s[2,2])
  sm = round(sm,2)
  return sm

sm = tensaomedia(s)
sm
```

```python
#Tensor Hidrostático
shid = np.dot(sm,np.eye(3))
shid
```

```python
#Tensor Distorcional
sij = s - shid
sij
```

```python
#Tensao media do tensor distorcional
smsij = tensaomedia(sij)
smsij
```
