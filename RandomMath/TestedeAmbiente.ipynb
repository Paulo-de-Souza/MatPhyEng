# Funções Trigonométricas



Dado um triângulo retângulo de lados A, B e C como apresentado na figura. Os lados A e B são conhecidos como catetos, sendo que A por estar ao lado do ângulo é denominado como **Cateto Adjacente** e B por estar em frente ao ângulo é chamado de **Cateto Oposto**, já o segmento C é conhecido como **Hipotenusa** e sempre será o maior lado do triângulo. 

![Figura 1](https://drive.google.com/uc?export=view&id=1MSpS-GrGhiySEr5Bbg0BBNSn2APkK0NY) 

Por definição temos que as funções seno, cosseno e tangente são respectivamente

$sen(\theta) = \frac{B}{C} \hspace{1cm}
cos(\theta) = \frac{A}{C} \hspace{1cm} tg(\theta) = \frac{sen(\theta)}{cos(\theta)} = \frac{B}{A}$



1.  Crie um programa que calcule o valor do seno, cosseno e tangente de 30 graus

```python
import numpy as np
thetadeg = 30;
thetarad = np.deg2rad(thetadeg)
s = np.sin(thetarad)
c = np.cos(thetarad)
t = np.tan(thetarad)
print("O valor do seno de {} é: {:.3f}\nO valor do cosseno de {} é: {:.3f}\nO valor da tangente de {} é: {:.3f}".format(thetadeg,s,thetadeg,c,thetadeg,t))
```



````python
import numpy as np
thetadeg = 30;
thetarad = np.deg2rad(thetadeg)
s = np.sin(thetarad)
c = np.cos(thetarad)
t = np.tan(thetarad)
print("O valor do seno de {} é: {:.3f}\nO valor do cosseno de {} é: {:.3f}\nO valor da tangente de {} é: {:.3f}".format(thetadeg,s,thetadeg,c,thetadeg,t))
````

2.   É sabido que o cateto adjacente e a hipotenusa de um triângulo retângulo medem respectivamente 20 e 35 centímetros, crie um programa que calcule o valor do cateto oposto e a tangente correspondente a esse triângulo; tome como base a figura da definição.



````python
import numpy as np
C = 35
A = 20
cosseno = A/C
#Calculando o ângulo do problema
theta = np.arccos(cosseno)
#Calculando o cateto oposto e a tangente
B = np.sin(theta)*C
T = np.tan(theta) 

print("O cateto oposto mede {:.2f} cm\nA tangente do problema vale {:.2f}".format(B,T))
````
